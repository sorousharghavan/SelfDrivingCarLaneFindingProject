# Processing individual images

In order to process individual images, first, we need to import the file list

```python
dirPath = "test_images/"
files = os.listdir(dirPath)
```

Now, we can simply loop thought the files and process each image

```python
for file in files:
    image = mpimg.imread(dirPath + file)
    image_output = process_image(image)
    show_image(image_output)
```
